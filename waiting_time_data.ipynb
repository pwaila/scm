```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import matplotlib.pyplot as plt
name='Popeyes'

q1=data[name].quantile(.25) 
print("Q1 = ",q1)
q2=data[name].quantile(.50) 
print("Q2 = ",q2)
q3=data[name].quantile(.75) 
print("Q3 = ",q3)
lowerLimit =q1-(1.5*(q3-q1))
print('lowerLimit',lowerLimit)
upperLimit =q3+(1.5*(q3-q1))
upperLimit
outliar=[x for x in data[name] if ((x > upperLimit) or (x<lowerLimit))]
print(outliar)

plt.style.use('classic')
%matplotlib inline


sns.set(style="whitegrid")
sns.boxplot(x=data[name])
```
